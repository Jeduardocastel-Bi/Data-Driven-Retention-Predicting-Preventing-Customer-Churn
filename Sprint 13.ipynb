Descripción del proyecto
La cadena de gimnasios Model Fitness está desarrollando una estrategia de interacción con clientes basada en datos analíticos.
Uno de los problemas más comunes que enfrentan los gimnasios y otros servicios es la pérdida de clientes. ¿Cómo descubres si un/a cliente ya no está contigo? Puedes calcular la pérdida en función de las personas que se deshacen de sus cuentas o no renuevan sus contratos. Sin embargo, a veces no es obvio que un/a cliente se haya ido: puede que se vaya de puntillas.
Los indicadores de pérdida varían de un campo a otro. Si un usuario o una usuaria compra en una tienda en línea con poca frecuencia, pero con regularidad, no se puede decir que ha huido. Pero si durante dos semanas no ha abierto un canal que se actualiza a diario, es motivo de preocupación: es posible que tu seguidor o seguidor/a se haya aburrido y te haya abandonado.
En el caso de un gimnasio, tiene sentido decir que un/a cliente se ha ido si no viene durante un mes. Por supuesto, es posible que estén en Cancún y retomen sus visitas cuando regresen, pero ese no es un caso típico. Por lo general, si un/a cliente se une, viene varias veces y luego desaparece, es poco probable que regrese.
Con el fin de combatir la cancelación, Model Fitness ha digitalizado varios de sus perfiles de clientes. Tu tarea consiste en analizarlos y elaborar una estrategia de retención de clientes.
Tienes que:


	• Aprender a predecir la probabilidad de pérdida (para el próximo mes) para cada cliente.
	• Elaborar retratos de usuarios típicos: selecciona los grupos más destacados y describe sus características principales.
	• Analizar los factores que más impactan la pérdida.
	• Sacar conclusiones básicas y elaborar recomendaciones para mejorar la atención al cliente:
		○ identificar a los grupos objetivo;
		○ sugerir medidas para reducir la rotación;

## Paso 2: Construir un modelo para predecir la cancelación de usuarios

### Crea un modelo de clasificación binaria para clientes donde la característica objetivo es la marcha del usuario o la usuaria el mes siguiente.

```python
# relacion entre las clases del dataset
print('Relación entre las clases:\n', df_gym['Churn'].value_counts())
```

```python
# Separar características (X) y variable objetivo (y)

X = df_gym.drop(columns=['Churn'])
y = df_gym['Churn']
```

### Divide los datos en conjuntos de entrenamiento y validación utilizando la función train_test_split().

```python
# dividir los datos en entrenamiento y prueba 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# crear objeto StandarScaler y aplicarlo al set de entrenamiento
scaler = StandardScaler()
X_train_st = scaler.fit_transform(X_train)

# aplicar la estandarizacion a la matriz del set de prueba 
X_test_st = scaler.transform(X_test)
```

<b>Comentario del revisor </b> <a class="tocSkip"></a>

Excelente, buen trabajo aplicando la estandarización a los datos y realizando el split de los mismos.     </div> 

### Entrena el modelo en el set de entrenamiento con dos métodos:
		○ regresión logística;
        ○ bosque aleatorio o Random Forest 

**Regresión Logística**

```python
# definir el algoritmo para el modelo de regresion logistica 

lr_model = LogisticRegression(random_state=0)
# entrenar al modelo 
lr_model.fit(X_train_st, y_train)
# utilizar el modelo entrenado para hacer previsiones
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:,1]
```

**Random Forest**

```python
# define el algoritmo para el nuevo modelo de random forest : Combina múltiples árboles de decisión 
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0)
# entrenar al modelo
rf_model.fit(X_train_st, y_train)
# utilizar el modelo entrenado para hacer previsiones
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
```

###  Evalúa la exactitud, precisión y recall para ambos modelos utilizando los datos de validación. Utilízalos para comparar los modelos. ¿Qué modelo dio mejores resultados?


```python
# definir funcion para mostrar metricas 
def print_metrics(y_true, y_pred, y_proba, title = 'Metricas de clasificación'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
```

**Evaluación de modelos**

¿Cómo elegir el mejor modelo?

Accuracy: Proporciona una visión general del rendimiento del modelo. Sin embargo, si las clases están desbalanceadas (por ejemplo, muchas más personas se quedan que cancelan), esta métrica puede ser engañosa.


Precision: Es crucial si el costo de predecir falsos positivos es alto. Por ejemplo, si prefieres no molestar a los clientes que no tienen riesgo de cancelar.


Recall: Es importante si el objetivo es identificar a la mayor cantidad posible de clientes que cancelarán, incluso si hay algunos falsos positivos.

***Para la regresión logística***

```python
print_metrics(y_test, lr_predictions, lr_probabilities , title='Métricas de regresión logística:')
```

***Para el Random Forest***

```python
print_metrics(y_test, rf_predictions, rf_probabilities, title='Métricas de random forest:')
```

**Al analizar los dos modelos se puede llegar a la conclusion que:**

Ambos modelos tienen una exactitud del 92%. Esto indica que, en general, ambos modelos hacen una correcta predicción en el 92% de los casos, lo que es muy bueno.

La precisión mide cuántos de los predichos como positivos (cancela) son realmente positivos. La regresión logística tiene una ligera ventaja con una precisión de 0.85 frente al bosque aleatorio con 0.84. Esto significa que la regresión logística tiene un pequeño margen de ventaja en evitar falsos positivos (clientes que no cancelan pero el modelo predice que lo harán).

El recall mide cuántos de los casos realmente positivos (los que cancelan) fueron identificados correctamente por el modelo. La regresión logística tiene un pequeño margen de ventaja con un recall de 0.83 frente al bosque aleatorio con 0.81, lo que sugiere que la regresión logística identifica un porcentaje ligeramente mayor de los clientes que realmente cancelan.


Aunque ambos modelos tienen un rendimiento similar, la regresión logística tiene una ligera ventaja en términos de precisión y recall. Si lo que te interesa es predecir de forma más precisa qué clientes cancelarán (y reducir los falsos positivos y falsos negativos), la regresión logística podría ser una mejor opción en este caso.

## Paso 3: Crear clústeres de usuarios/as

### Deja de lado la columna con datos sobre la cancelación e identifica los clústeres de objetos (usuarios/as)

columnas que describen el comportamiento de los usuarios en el gimnasio:

Avg_class_frequency_total: Frecuencia promedio de visitas a clases a lo largo de toda la vida del cliente.

Avg_class_frequency_current_month: Frecuencia promedio de visitas a clases durante el mes actual.

Avg_additional_charges_total: Monto total de dinero gastado por el usuario en otros servicios del gimnasio (como productos deportivos, cafetería, masajes, etc.).

```python
# seleccionar columnas del comportamiento de usuario en el gym 

avg_columns = df_gym[['Avg_class_frequency_total', 
                       'Avg_class_frequency_current_month', 
                       'Avg_additional_charges_total']]
```

```python
avg_columns
```

###  Estandariza los datos.

Estandarizar significa transformar los datos de manera que cada característica (columna) tenga una media de 0 y una desviación estándar de 1. Esto permite que todas las características tengan el mismo peso en el análisis de clústeres, y no se vea afectado por las unidades o rangos de los datos.

```python
# estandarizar los datos 

sc = StandardScaler()
x_sc_avg = sc.fit_transform(avg_columns)
```

###  Utiliza la función linkage() para crear una matriz de distancias basada en la matriz de características estandarizada y trazar un dendrograma

```python
# funcion linkage
linked_avg = linkage(x_sc_avg, method='average')
```

```python
# Trazar dendrograma para las columnas de promedio
plt.figure(figsize=(12, 8))
dendrogram(
    linked_avg,
    truncate_mode='level',  # Mostrar solo un número de niveles del árbol
    p=6,  # Número de niveles principales que quieres mostrar
    show_leaf_counts=True,  # Mostrar el conteo de datos en cada rama
    leaf_rotation=90,
    leaf_font_size=10
)

plt.title('Dendrograma basado en comportamiento (Avg)')
plt.ylabel('Distancia')
plt.xlabel('Grupos principales')
plt.show()
```

El dendrograma muestra dos grandes clusters principales, lo que indica que los datos pueden clasificarse en dos grupos distintos si cortamos a una altura razonable.
Los subclusters dentro de cada grupo representan agrupaciones más pequeñas de datos que tienen un comportamiento o características más similares entre sí.

### Entrena el modelo de clustering con el algortimo K-means y predice los clústeres de clientes.

```python
# definir el modelo
model_cluster = KMeans(n_clusters = 5)

# predice los clusters para las observaciones 
labels = model_cluster.fit_predict(x_sc_avg)
```

###  Traza distribuciones de características para los clústeres. ¿Notas algo?
